# Tipos en TypeScript

춰Por supuesto! Entender los tipos de datos en TypeScript es fundamental, como su nombre lo sugiere esto es  **mucho muy importante 游땕**.
Los tipos en TypeScript se pueden definir como constantes (const) y variables (let) de la siguiente manera:


```typescript
//programa_1.ts
// Imprimir en consola
console.log("Hola mundo");

// Tipos b치sicos
let numero: number = 10;
let texto: string = "Hola";
let booleano: boolean = true;

// Imprimir tipos b치sicos
console.log(`Numero: ${numero}, Texto: ${texto}, Booleano: ${booleano}`);

// Tipos compuestos
let arreglo: number[] = [1, 2, 3];
let tupla: [string, number] = ["Hola", 10];
let objeto: { nombre: string, edad: number } = { nombre: "Juan", edad: 30 };

// Imprimir tipos compuestos
console.log(`Arreglo: ${arreglo}, Tupla: ${tupla}, Objeto: ${objeto}\n`);

// Tipos no tipos: null  any
let nulo: null = null;
let cualquiera: any = "Hola";

//el tipo que yo yamo el verdadero tipo de undefined es similar void
let indefinido: undefined = undefined;


console.log(`Nulo: ${nulo}, Indefinido: ${indefinido}, Cualquiera: ${cualquiera}\n`);

// Declaraci칩n de variable sin tipo
let sinTipo = "Hola";

// Imprimir variable sin tipo
console.log(`Sin tipo: ${sinTipo}\n`);

// Constantes
const constante: string = "Hola";

// Imprimir constante
console.log(`Constante: ${constante}\n`);

// Cambio de valor de variable
// sinTipo = "Adios"; //error 


// Clases de los primitivos
console.log(`Tipo de numero: ${typeof numero}`);
console.log(`Tipo de texto: ${typeof texto}`);
console.log(`Tipo de booleano: ${typeof booleano}`);
console.log(`Tipo de arreglo: ${typeof arreglo}`);
console.log(`Tipo de tupla: ${typeof tupla}`);
console.log(`Tipo de objeto: ${typeof objeto}`);
console.log(`Tipo de nulo: ${typeof nulo}`);
console.log(`Tipo de indefinido: ${typeof indefinido}`);
console.log(`Tipo de cualquiera: ${typeof cualquiera}`);
console.log(`Tipo de sinTipo: ${typeof sinTipo}`);
console.log(`Tipo de constante: ${typeof constante}`);

//tambien se puede concatenar la salidas de los console.log
console.log(`Tipo de numero: ${typeof numero}, Tipo de texto: ${typeof texto}, Tipo de booleano: ${typeof booleano}, Tipo de arreglo: ${typeof arreglo}, Tipo de tupla: ${typeof tupla}, Tipo de objeto: ${typeof objeto}, Tipo de nulo: ${typeof nulo}, Tipo de indefinido: ${typeof indefinido}, Tipo de cualquiera: ${typeof cualquiera}, Tipo de sinTipo: ${typeof sinTipo}, Tipo de constante: ${typeof constante}`);
//con el operador + tambien se puede concatenar
console.log("Tipo de numero: " + typeof numero + ",\n Tipo de texto: " + typeof texto + ",\n Tipo de booleano: " + typeof booleano + ",\n Tipo de arreglo: " + typeof arreglo + ",\n Tipo de tupla: " + typeof tupla + ",\n Tipo de objeto: " + typeof objeto + ",\n Tipo de nulo: " + typeof nulo + ",\n Tipo de indefinido: " + typeof indefinido + ",\n Tipo de cualquiera: " + typeof cualquiera + ",\n Tipo de sinTipo: " + typeof sinTipo + ",\n Tipo de constante: " + typeof constante);
```


1. **Imprimir en consola:** Se utiliza `console.log()` para mostrar "Hola mundo" en la consola.

2. **Tipos b치sicos:** Se declaran variables con tipos b치sicos (`number`, `string`, `boolean`).

3. **Tipos compuestos:** Se declaran variables con tipos compuestos (`number[]`, `[string, number]`, `{ nombre: string, edad: number }`).

4. **Objetos** Se declaran variables con tipos especiales (`null`, `any`) y se imprimen.

5. **Variable sin tipo:** Se declara una variable sin especificar su tipo, se infiere como `string` y se imprime.

6. **Constantes:** Se declara una constante y se intenta modificar (comentado) para mostrar que las constantes no pueden ser reasignadas.

7. **Clases de los primitivos:** Se usa `typeof` para obtener el tipo de las variables y se imprimen.

8. **El tipo undefined:** como podemos ver el typo undefined no esta definido

:::{important} **`var` deprecated**  
Usualmente encontrar치s `var` en muchos tutoriales; sin embargo, est치 en desuso y se debe reemplazar por `let` o `const`.
